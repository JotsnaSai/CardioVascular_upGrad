```python
corr = cardio[['age', 'diabetes']].corr()
corr.style.background_gradient(cmap='coolwarm')
```

So from the above correlation between age and diabetes we can say that there is no relation between these two.        
In the correlation, we will consider only if the value is above 0.5. Here the value is 0.1 in diabetes.          
Now a days even small kids are getting diabetes. So it is proved that there is nothing to do with age.

```python
cardio.drop("id",axis=1,inplace=True)
cardio
```

Dropped id column because it is of no use

```python
cardio.drop("education",axis=1,inplace=True)
cardio
```

Above I have removed the education column because the data was not appropriate.

```python
cardio.isnull().sum()
```

```python
cardio['cigsPerDay'].fillna(int(cardio['cigsPerDay'].mean()), inplace=True)
```

```python
cardio = cardio.fillna({'BPMeds' : '0'})
```

```python
cardio['totChol'].fillna(int(cardio['totChol'].median()), inplace=True)
```

```python
cardio['BMI'].fillna(int(cardio['BMI'].mean()), inplace=True)
```

```python
cardio['glucose'].fillna(int(cardio['glucose'].mean()), inplace=True)
```

```python
cardio.isnull().sum()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (13, 5))
plt.subplots_adjust(wspace = 0.4)

ax1 = sns.scatterplot(data = cardio, x = 'BMI', y = 'sysBP', hue = 'prevalentStroke', ax = axes.flat[0])
ax1.set_title('weight and ap_hi')

ax2 = sns.scatterplot(data = cardio, x = 'BMI', y = 'diaBP', hue = 'prevalentStroke', ax = axes.flat[1])
ax2.set_title('weight and ap_lo')

sns.despine();
```

```python
print("Cardiovascular Disease Presence in Men and Women (in Percentage):")
print(round(cardio.groupby('sex')['prevalentStroke'].value_counts() / cardio.groupby('sex')['prevalentStroke'].count() * 100, 2))
```

```python
print("Percentage of men and women who are smoking")
print(round(cardio.groupby('sex')['is_smoking'].value_counts() / cardio.groupby('sex')['is_smoking'].count() * 100, 2))
```

```python
numeric_cols = cardio.select_dtypes(include = np.number).columns.tolist()
print("Numerical Columns:", numeric_cols)

categorical_cols = cardio.select_dtypes(include = 'object').columns.tolist()
print("Categorical Columns:", categorical_cols)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (18, 8))

fig.suptitle('Categorical Columns Based on Cardio', y = 0.95, fontsize = 16)
plt.subplots_adjust(hspace = 0.4, wspace = 0.4)

for pos, name in enumerate(categorical_cols[:-1]):
  sns.countplot(data = cardio, x = name, hue = 'prevalentStroke', ax = axes.flat[pos], alpha = 0.5);

sns.despine()
```
